```python
import os
import pandas as pd
import numpy as np
import math
```

```python
url = "/content/NuevoDataset.csv"
```

```python
df = pd.read_csv(url)
```

```python
df.describe()
```

```python
##Feature: Abandono
abandono1= df['2P']==0
abandono2= df['Requisito']==0
abandono3= df['1P']<12
abandono4= df['NF']==0

abandono1=abandono1.replace([True,False],[1,0])
abandono2=abandono2.replace([True,False],[1,0])
abandono3=abandono3.replace([True,False],[1,0])
abandono4=abandono4.replace([True,False],[1,0])

abandonoSum1 = abandono1+abandono2
abandonoSum2 = abandono3+abandono4
abandonoSum = abandonoSum1 + [1 if x == 2 else 0 for x in abandonoSum2]
abandono = [1 if x > 0 else 0 for x in abandonoSum]

df.insert(21,"Abandono",abandono,True)
```

```python
##Feature: Sumatoria de parciales+taller
Firma = df['1P'] + df['2P'] + df['Taller']
df.insert(14,"SumProceso",Firma)
```

```python
##Feature: Veces recursada
df = df.sort_values(by=['Idalumno','Anho'])
df['Veces recursada'] = df.groupby(['Idalumno']).cumcount()
```

```python
 #Creacion de un nuevo dataset con el dataframe final
df.to_csv("Estatica_Dataset.csv",index=False)
```
